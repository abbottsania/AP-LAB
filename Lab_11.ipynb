# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
import sys

def tsp_with_path(graph):
    n = len(graph)
    dp = [[(float('inf'), -1)] * (1 << n) for _ in range(n)]
    def solve(city, mask):
        if mask == (1 << n) - 1:
            return graph[city][0], 0

        if dp[city][mask][0] != float('inf'):
            return dp[city][mask]

        min_cost = float('inf')
        next_optimal_city = -1

        for next_c in range(n):
            if not (mask & (1 << next_c)):
                cost_to_next, _ = solve(next_c, mask | (1 << next_c))
                current_path_cost = graph[city][next_c] + cost_to_next

                if current_path_cost < min_cost:
                    min_cost = current_path_cost
                    next_optimal_city = next_c

        dp[city][mask] = (min_cost, next_optimal_city)
        return dp[city][mask]
    min_total_cost, _ = solve(0, 1 << 0)

    path = []
    current_city = 0
    current_mask = 1 << 0

    while True:
        path.append(current_city)

        if current_mask == (1 << n) - 1:
            break

        _, next_city = dp[current_city][current_mask]

        if next_city == -1:
            print("Error: Could not reconstruct path fully.")
            break

        current_city = next_city
        current_mask |= (1 << current_city)

    path.append(0)

    return min_total_cost, path

def get_user_graph_input():
    while True:
        try:
            num_cities = int(input("Enter the number of cities: "))
            if num_cities < 2:
                print("Please enter at least 2 cities for a meaningful TSP.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer for the number of cities.")

    graph = [[0] * num_cities for _ in range(num_cities)]

    print("\nEnter the distance matrix:")
    print("For each cell [i][j], enter the distance from city i to city j.")
    print("Enter '0' for distance from a city to itself.")
    print("For example, if distances are symmetric (i.e., dist(A,B) == dist(B,A)), you will still enter both values.")

    for i in range(num_cities):
        for j in range(num_cities):
            if i == j:
                graph[i][j] = 0
                continue
            while True:
                try:
                    distance = int(input(f"Distance from city {i} to city {j}: "))
                    if distance < 0:
                        print("Distance cannot be negative. Please enter a non-negative integer.")
                        continue
                    graph[i][j] = distance
                    break
                except ValueError:
                    print("Invalid input. Please enter an integer for distance.")
    return graph, num_cities

print("--- Traveling Salesman Problem Solver with Path Reconstruction ---")
user_graph, num_cities_input = get_user_graph_input()
print(f"\nCalculating TSP for {num_cities_input} cities...")
min_cost, path = tsp_with_path(user_graph)

print("\nMinimum traveling cost:", min_cost)
path_str = " -> ".join(map(str, path))
print("Optimal path:", path_str)
```
