# Problem Definition
Given a set of
n
n items, each with:
Value vi
Weight wi
A knapsack with maximum capacity **W**

Goal: maximize total value that can be put in the knapsack.
Constraint: You may take fractions of items.

Greedy Strategy:

*  Sort items by decreasing ratio vi/wi
*  Select items fully until knapsack is full.
*  Take fraction of the next item if capacity remains.

```python
def fractional_knapsack(capacity, items):
    """Solves the Fractional Knapsack Problem using a greedy approach.

    Args:
        capacity (int or float): The maximum capacity of the knapsack.
        items (list of tuples): A list of items, where each item is a tuple
                                (value, weight).

    Returns:
        tuple: A tuple containing:
            - float: The maximum total value that can be put in the knapsack.
            - list: A list of tuples (item_value, item_weight, fraction_taken)
                    representing the items taken and their fractions.
    """

    indexed_items = []
    for i, (value, weight) in enumerate(items):
        if weight > 0:
            ratio = value / weight
        else:
            ratio = 0
        indexed_items.append((ratio, value, weight, i))
    indexed_items.sort(key=lambda x: x[0], reverse=True)

    total_value = 0
    knapsack_contents = []
    remaining_capacity = capacity

    for ratio, value, weight, original_index in indexed_items:
        if remaining_capacity == 0:
            break

        if weight <= remaining_capacity:
            total_value += value
            remaining_capacity -= weight
            knapsack_contents.append((value, weight, 1.0))
        else:
            fraction = remaining_capacity / weight
            total_value += value * fraction
            knapsack_contents.append((value, weight, fraction))
            remaining_capacity = 0
            break

    return total_value, knapsack_contents
if __name__ == "__main__":
    # Items: (value, weight)
    items = [(60, 10), (100, 20), (120, 30)]
    capacity = 50

    max_value, contents = fractional_knapsack(capacity, items)

    print(f"Knapsack Capacity: {capacity}")
    print(f"Items available: {items}")
    print(f"\nMaximum total value: {max_value:.2f}")
    print("Items taken:")
    for val, wt, frac in contents:
        print(f"  - Value: {val}, Weight: {wt}, Fraction taken: {frac:.2f}")

    print("\n--- Another Example ---")
    items2 = [(500, 100), (300, 50), (400, 70)]
    capacity2 = 120

    max_value2, contents2 = fractional_knapsack(capacity2, items2)

    print(f"Knapsack Capacity: {capacity2}")
    print(f"Items available: {items2}")
    print(f"\nMaximum total value: {max_value2:.2f}")
    print("Items taken:")
    for val, wt, frac in contents2:
        print(f"  - Value: {val}, Weight: {wt}, Fraction taken: {frac:.2f}")
```

In this lab, we implement Huffman Coding, a greedy algorithm used for lossless data compression.
It assigns variable-length binary codes to input symbols based on their frequencies, ensuring that more frequent symbols receive shorter codes and less frequent symbols receive longer codes.

This algorithm constructs a binary tree (Huffman Tree) where each leaf node represents a character, and the path from the root to a leaf gives the character’s binary code.

Use:
Huffman Tree Construction using a Min-Priority Queue (Min-Heap).
Code generation for each character.
Encoding and decoding using the generated Huffman codes.
Complexity analysis and comparison with fixed-length encoding.

# Pseudocode:
1. Create a leaf node for each character and insert all nodes into a
2. min-heap by frequency.
   *  While the heap has more than one node:
   * Extract two nodes with the lowest frequencies.
   * Create a new internal node with frequency = sum of the two.
   *  Set the two extracted nodes as left and right children.
   *  Insert the new node back into the min-heap.
3. The remaining node is the root of the Huffman Tree.

4. Traverse the tree:
   * Assign 0 for left edge and 1 for right edge.
   * The resulting paths from root to leaves give Huffman codes.

```python
import heapq

class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None
    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(frequencies):
    """
    Builds a Huffman tree from a dictionary of character frequencies.

    Args:
        frequencies (dict): A dictionary where keys are characters and values are their frequencies.

    Returns:
        Node: The root node of the Huffman tree.
    """
    priority_queue = []
    for char, freq in frequencies.items():
        heapq.heappush(priority_queue, Node(char, freq))
    while len(priority_queue) > 1:
        left = heapq.heappop(priority_queue)
        right = heapq.heappop(priority_queue)
        merged_freq = left.freq + right.freq
        merged_node = Node(None, merged_freq)
        merged_node.left = left
        merged_node.right = right


        heapq.heappush(priority_queue, merged_node)


    return priority_queue[0]
if __name__ == "__main__":
    frequencies = {
        'a': 5,
        'b': 9,
        'c': 12,
        'd': 13,
        'e': 16,
        'f': 45
    }

    print("Building Huffman Tree for frequencies:", frequencies)
    huffman_tree_root = build_huffman_tree(frequencies)
    print("Huffman Tree built successfully (root node returned).")
```

The `build_huffman_tree` function takes a dictionary of character frequencies and returns the root of the constructed Huffman tree. The `Node` class helps represent each node in the tree. The next step would be to generate the Huffman codes by traversing this tree.
